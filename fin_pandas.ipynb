```python
import requests
import re
from bs4 import BeautifulSoup
from collections import defaultdict
import pandas as pd
raw_data= defaultdict(list)
counter=0
links=[]
```

```python
# Extract all pages
page_pre ="http://aajtak.intoday.in/photos/sports.html/page/"
page_list=[]
for i in range(200,600,1):
    page_list.append(page_pre+str(i))
```

```python
def make_links(page_param):
    global links
    page = requests.get(page_param)
    soup = BeautifulSoup(page.content,'html.parser')
    for i in soup.find_all('a',href=re.compile("/gallery/")):
        links.append("http://aajtak.intoday.in"+i["href"])
```

```python
for i in page_list :
    make_links(i)
print len(links)
filtered_links= set(links)
```

```python
print len(filtered_links)
```

```python
x = list(filtered_links)
raw_data["links"]=x
df = pd.DataFrame(raw_data,columns=["links"])
```
